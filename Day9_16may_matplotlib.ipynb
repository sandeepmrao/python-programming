#### scatter plot:


```python
dft=pd.read_csv('train.csv')
```

```python
dft.columns
```

```python
x=dft['Survived'] #x-axis of plot
y=dft['Age']   # y axis 
plt.scatter(x,y)
plt.xlabel('Survived')   # labelling the x axis
plt.ylabel('Age')
plt.title('Age vs Survived in Titanic')
plt.show()
```

```python
x=dft['Sex'] #x-axis of plot
y=dft['Age']   # y axis 
plt.scatter(x[:500],y[:500],marker='*',color='g',s=100,edgecolor='r')
plt.xlabel('Sex')   # labelling the x axis
plt.ylabel('Age')
plt.title('Age vs Sex in Titanic')
plt.grid()  # for drawing grid lines
plt.show()
```

```python
x=dft['Cabin'] #x-axis of plot
y=dft['Survived']   # y axis 
plt.scatter(x,y,marker='+',color='b',edgecolor='r')
plt.xlabel('Cabin')   # labelling the x axis
plt.ylabel('Survived')
plt.title('Cabin vs Surv. in Titanic')
plt.grid()
plt.show()
```

#### line graph: plot()

```python
x=np.arange(1,11)
y=x**2
z=x**3
plt.plot(x,y,label='Quadratic',marker='+')
plt.plot(x,z,label='cubic',marker='*')
plt.legend()
plt.xlabel('Whole nos.')
plt.ylabel('quadm & cube')
plt.grid()
plt.axis([1,15,1,1500])
plt.show()
```

```python
print(plt.style.available)
```

```python
x=np.arange(1,11)
y=x**2
z=x**3
plt.style.use('seaborn-dark')
plt.plot(x,y,label='Quadratic')
plt.plot(x,z,label='cubic')
plt.legend()
plt.xlabel('Whole nos.')
plt.ylabel('quadm & cube')
plt.grid()
plt.axis([1,15,1,1500])
plt.show()
```

```python
x=np.arange(1,11)
y=x**2
z=x**3
plt.plot(x,y,label='Quadratic',c='b')
plt.plot(x,z,label='cubic',c='r')
plt.legend()
plt.fill_between(x,z,600,where=z<400)
```
